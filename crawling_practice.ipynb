- HTML 문서 내에 ID가 mw-content-text인 태그내의 내용을 출력해주세요.

### 실습 4
- https://ko.wikipedia.org/wiki/포털:요즘_화제
- 사이트에서 주요뉴스 내용과 일자별 뉴스들을 리스트로 정리해주세요
- 결과 : [‘주요 뉴스’, '8월 25일, 파키스탄에서 일어난 홍수로 1,000명 이상의 주민이 사망하고, 가축 70,000마리 이상이 죽었다.’, ...,
'Current events of 2022년 8월 29일\xa0(2022-08-29) (월요일)', ‘대한민국의 0시 기준 누적 \u200b확진자 수가 23,026,960명으로
집계되었다. 전날 0시 대비 43,142명(국내 42,782, 해외유입 360)이 늘었다.’, ...]

```python
url = "https://ko.wikipedia.org/wiki/포털:요즘_화제"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')
table_tags = soup.select('.vevent')           

result = ['주요 뉴스']

li_tags = table_tags[0].select('li')        # 주요 뉴스 항목
for li_tag in li_tags:
    result.append(li_tag.text.strip())

for table_tag in table_tags[1:-3]:          # 일자별 뉴스
    day = table_tag.select_one('.summary').text.strip()    # 일자
    content = table_tag.select_one('.description').text.strip()    # 내용
    result.append([day, content])
result
```

### 실전 1
- https://finance.naver.com/sise/sise_quant.nhn
1. 종목명과 현재가를 크롤링해주세요.
2. 전일대비 상승한 종목만 종목명, 현재가, 전일비를 크롤링해주세요.

#### 1번

```python
# 1. slicing
url = "https://finance.naver.com/sise/sise_quant.nhn"
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

tr_tags = soup.select('.type_2 tr')         # 초기 html에는 Javascript 동작 실행 전이므로 tbody 존재 x -> 자식 셀렉터로도 가능
for i in range(20):
    targets = tr_tags[2:][8*i:8*i+5]
    titles = [t.select_one('.tltle').text.strip() for t in targets]
    print(titles)

# 2. try-except
result = []
for tr_tag in tr_tags[2:]:
    try:
        # print(tr_tag.select_one('.tltle').text.strip())
        tmp = {
            '종목명': tr_tag.select_one('.tltle').text.strip(),
            '현재가': tr_tag.select_one('.number').text.strip(),
        }
        result.append(tmp)
    except:
        pass

# 3. if-else
for tr_tag in tr_tags[2:]:
    if tr_tag.select_one('.tltle'):
        print(tr_tag.select_one('.tltle').text.strip())
    else:
        pass
```

```python
# 2. try-except
result = []
tr_tags = soup.select('.type_2 tr')
for tr_tag in tr_tags[2:]:
    try:
        # print(tr_tag.select_one('.tltle').text.strip())
        tmp = {
            '종목명': tr_tag.select_one('.tltle').text.strip(),
            '현재가': tr_tag.select_one('.number').text.strip(),
        }
        result.append(tmp)
    except:
        pass
result
```

#### 2번

```python
# try-except
up_stocks = []
tr_tags = soup.select('.type_2 tr')
for tr_tag in tr_tags[2:]:
    try:
        # 이미지 체크 -> alt 속성 체크
        if tr_tag.select_one('img'):
            # print(tr_tag.select_one('img').attrs['src'])
            if 'ico_up' in tr_tag.select_one('img').attrs['src']:
                up_stocks.append({
                    '종목명': tr_tag.select_one('.tltle').text.strip(),
                    '현재가': tr_tag.select_one('.number').text.strip(),
                    '전일비': tr_tag.select('.number')[1].text.strip()
                })
    except:
        pass
up_stocks
```

### 실전 2
- https://store.musinsa.com/app/items/lists/001
1. 무신사 쇼핑몰 상의(TOP) 카테고리 첫페이지 제품들의 브랜드명, 제품명, 원래가격, 할인가격을 크롤링해주세요.
2. 성공하신분들은 총 10페이지까지 크롤링해주세요.

#### 1번

```python
url = "https://www.musinsa.com/categories/item/001"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

tops = []
top_lists = soup.select('#searchList .article_info')
for top in top_lists:
    tops.append({
        '브랜드명': top.select_one('.item_title').text.strip(),
        '제품명': top.select_one('.list_info').text.strip(),
        '원래가격': top.select_one('.price').text.strip().split('\n')[0],
        '할인가격': top.select_one('.price').text.strip().split('\n')[-1].strip()
    })

tops
```

```python
# 1페이지: "https://www.musinsa.com/categories/item/001?d_cat_cd=001&brand=&list_kind=small&sort=pop_category&sub_sort=&page=1&display_cnt=90&group_sale=&exclusive_yn=&sale_goods=&timesale_yn=&ex_soldout=&plusDeliveryYn=&kids=&color=&price1=&price2=&shoeSizeOption=&tags=&campaign_id=&includeKeywords=&measure="
# 2페이지: "https://www.musinsa.com/categories/item/001?d_cat_cd=001&brand=&list_kind=small&sort=pop_category&sub_sort=&page=2&display_cnt=90&group_sale=&exclusive_yn=&sale_goods=&timesale_yn=&ex_soldout=&plusDeliveryYn=&kids=&color=&price1=&price2=&shoeSizeOption=&tags=&campaign_id=&includeKeywords=&measure="

# page를 변수로 받아 반복문 실행
tops = []
for i in range(1, 11):
    url = f"https://www.musinsa.com/categories/item/001?d_cat_cd=001&brand=&list_kind=small&sort=pop_category&sub_sort=&page={i}"
    print(url)
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    top_lists = soup.select('#searchList .article_info')
    for top in top_lists:
        tops.append({
            '브랜드명': top.select_one('.item_title').text.strip(),
            '제품명': top.select_one('.list_info').text.strip(),
            '원래가격': top.select_one('.price').text.strip().split('\n')[0],
            '할인가격': top.select_one('.price').text.strip().split('\n')[-1].strip()
        })

tops
```
