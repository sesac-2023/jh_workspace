### 실전 1
- 제목과 썸네일을 딕셔너리로 만들어서 리스트에 담아주세요.
- https://comic.naver.com/webtoon/weekday?order=User

```python
url = 'https://comic.naver.com/api/webtoon/titlelist/weekday?order=view'

response = requests.get(url)
data = json.loads(response.text)

webtoons = []
for key in data['titleListMap']:
    for webtoon in data['titleListMap'][key]:
        webtoons.append({
            '제목': webtoon['titleName'],
            '썸네일': webtoon['thumbnailUrl']
        })

webtoons
```

### 실전 3
- 회사명, 회사 설명, 채용공고 (회사별 여러 개)를 총 10페이지를 크롤링 해주세요.
- https://www.rocketpunch.com/jobs

```python
url = 'https://www.rocketpunch.com/api/jobs/template?page=2'

response = requests.get(url)
data = json.loads(response.text)
html = data['data']['template']     # html로 구성되어 있기 때문에 BeautifulSoup 사용

soup = BeautifulSoup(html, 'html.parser')
company_list = soup.select('.company.item')

rocketpunch = []
for company in company_list:
    rocketpunch.append({
        "회사명": company.select_one('.company-name').text.strip(),
        "회사설명": company.select_one('.description').text.strip(),
        "채용공고": [(job.text.strip(), job.select_one('a').attrs['href']) for job in company.select('.job-name')]
        # "채용공고": [(job.text.strip(), job.attrs['href']) for job in company.select('.job-name > a:first-child')]
    })

rocketpunch
```

```python
def crawl_rocketpunch(page):
    response = requests.get(url.format(page))
    data = json.loads(response.text)
    html = data['data']['template']     # html로 구성되어 있기 때문에 BeautifulSoup 사용

    soup = BeautifulSoup(html, 'html.parser')
    company_list = soup.select('.company.item')

    rocketpunch = []
    for company in company_list:
        rocketpunch.append({
            "회사명": company.select_one('.company-name').text.strip(),
            "회사설명": company.select_one('.description').text.strip(),
            "채용공고": [(job.text.strip(), job.select_one('a').attrs['href']) for job in company.select('.job-name')]
            # "채용공고": [(job.text.strip(), job.attrs['href']) for job in company.select('.job-name > a:first-child')]
        })

    return rocketpunch

rocketpunch = []
for page in range(1, 11):
    rocketpunch.extend(crawl_rocketpunch(page))

rocketpunch
```

### 실전 4
- 일별 시세 탭에 있는 데이터를 총 10페이지까지 모두 크롤링 해주세요.
- https://finance.naver.com/item/sise.naver?code=005930

```python
url = 'https://finance.naver.com/item/sise_day.naver?code=005930&page=1'
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
    }

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

stocks = []
for tr_tag in soup.select('tr'):
    if tr_tag.select_one('.num'):
        stocks.append([td.text.strip() for td in tr_tag.select('td')])
    else:
        pass

stocks
```

```python
def samsung_electronics_crawling(page):
    response = requests.get('https://finance.naver.com/item/sise_day.naver?code=005930&page={}'.format(page), headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    stocks = []
    for tr_tag in soup.select('tr'):
        if tr_tag.select_one('.num'):
            stocks.append([td.text.strip() for td in tr_tag.select('td')])
        else:
            pass
    return stocks

stocks = []
for page in range(1, 11):
    stocks.extend(samsung_electronics_crawling(page))

stocks
```
