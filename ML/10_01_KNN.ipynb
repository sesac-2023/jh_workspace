```python
import numpy as np
#
def euclidean_distance(a:np.array, b:np.array):
    dist = 0
    # 계산
    # x1-y1 제곱 + x2-y2 제곱 + x3-y3 제곱 + ... + xn-yn 제곱 -> sqrt

    return np.sum((a-b)**2)**0.5

def cosine_similarity(a:np.array, b:np.array):
    similarity = 0
    # 계산
    similarity = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
    # ab 내적 / (a크기 * b크기)
    return similarity
```

```python
result_dist = []
result_sim = []
# result[0] => 0번째 index 데이터를 기준으로 0~177 각각 dist 계산한 값들이 들어가야됩니다
# result[0] = [0, 0.3, 0.7, ... 177]
# result[1] = [0.3, 0, ... , ]

for i in range(len(wine_df)): # 기준 데이터
    temp_dist = []
    temp_sim = []
    for j in range(len(wine_df)): # 비교대상
        base_point = wine_df.iloc[i].values
        target_point = wine_df.iloc[j].values
        # distance 계산
        dist = euclidean_distance(base_point, target_point)
        temp_dist.append(dist)
        # similarity 계산
        simm = cosine_similarity(base_point, target_point)
        temp_sim.append(simm)

    result_dist.append(temp_dist)
    result_sim.append(temp_sim)
```
